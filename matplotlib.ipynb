```python
plt.figure(figsize=(12,4))
plt.subplot(2,2,1)  # sybplot(nbre ligne, nbr colonne, numero de figure)
plt.plot(x,y, c='red')

plt.subplot(2,2,2)
plt.plot(x,y, c='blue')

plt.subplot(2,2,3)
plt.plot(x,y, c='blue')

plt.subplot(2,2,4)
plt.plot(x,y, c='blue')
plt.show()     # display toute la figure
```

```python
# OOP method
fig, ax= plt.subplots()   # fig est une objet && ax est un tableau des objets
ax.plot(x,y)  #=== ax[0].plot(x,y)
plt.show()

fig, ax= plt.subplots(1,2,sharex=True)
ax[0].plot(x,y)
ax[1].plot(x,y)
plt.show()   #afficher figure
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
iris = load_iris()
x = iris.data    # this dataset contain 5 feautures, bt it represent juste 4 of them :  sepal length, sepal width, petal length, and petal width
y = iris.target  # return the 5eme feature "class" :
                  #0: Setosa
                  # 1: Versicolour
                  # 2: Virginica

names = list(iris.target_names)
print(names)
print(f'x contient {x.shape[0]} exmples et {x.shape[1]} variables')
print(f'il y a {np.unique(y).size} classes')

#plt.scatter(x[:,0], y[:,0])   # erreur car this annotation y[..,..] is done juste for multi-dimensional array
plt.scatter(x[:,0], y)

# x=np.array([[1], [2],[3], [4]])
# y= np.array([1,5,3,4])
# plt.scatter(x,y)
# plt.show()
```
