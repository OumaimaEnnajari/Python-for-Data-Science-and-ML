```python
#============ exercice sur image en couleur  ==============

from scipy import misc  #misc is a module
import matplotlib.pyplot as plt # pyplot is a module

face = misc.face()  # load this image= un array de 3 dimensions , chaque pixiel est un tablea de shape= (3, ) [r,g,b]
plt.imshow(face)    # afficher image mais ne s'affiche pas a l'ecran
plt.show()          # afficher image sur ecran
face.shape
```

```python
#============ exercice sur image en gris  ==============

from scipy import misc  #misc is a module
import matplotlib.pyplot as plt # pyplot is a modulec

face = misc.face(gray=True) # load this image= un array de 2 dimensions chaque pixel est une valeur scalaire entre 0 et 255
plt.imshow(face, cmap= plt.cm.gray)    # afficher image mais ne s'affiche pas a l'ecran
plt.show()                              # afficher image sur ecran
print(face)
```

```python
from scipy.interpolate import interp1d
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.linspace(0,10,10)
y=x*2
plt.scatter(x,y)
```

```python
#=============== Interpolation ==================
f=interp1d(x,y,kind='linear')  #kind='cubic' ...
new_x= np.linspace(0,10,50)
new_y= f(new_x)
plt.scatter(x,y)
plt.scatter(new_x, new_y )
plt.show()
```
