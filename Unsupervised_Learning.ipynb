```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# Génération de données
X, y = make_blobs(n_samples=100, centers=3, cluster_std=0.4, random_state=0)
plt.scatter(X[:,0], X[:,1])
```

```python
estimator= KMeans(n_clusters= 3)
estimator.fit(X)
estimator.labels_ # return the number of cluster where each data set point is belong to
estimator.predict(X)

plt.scatter(X[:, 0], X[:, 1], c= estimator.labels_)

estimator.cluster_centers_
plt.scatter(estimator.cluster_centers_[:, 0], estimator.cluster_centers_[:, 1], c= 'r' )
```

```python
print(estimator.score(X))  #=== is equal to -Inertia value

print(estimator.inertia_)
```

```python
#=============== elbow technique ======================

inertia=[]
for k in range(1, 20):
  model = KMeans(n_clusters=k)
  model.fit(X)
  inertia.append(model.inertia_)

plt.plot(range(1, 20) , inertia)
```
